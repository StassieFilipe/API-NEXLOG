```python
import dask.dataframe as dd
import os

# 1. REMOVIDA A BARRA NO FINAL DOS ARQUIVOS
# Certifique-se de que o caminho do arquivo2 está correto (mudei de 'sua_pata' para 'API_2LOG')
arquivo1 = '/content/drive/MyDrive/API_2LOG/liquidos_gasosos.csv'
arquivo2 = '/content/drive/MyDrive/API_2LOG/solidos.csv'

lista_especifica = [arquivo1, arquivo2]

# Verificação de existência
existentes = [f for f in lista_especifica if os.path.exists(f)]

if len(existentes) == 2:
    # O Dask lê apenas os dois
    df_unificado = dd.read_csv(existentes, sep=None, engine='python')

    # IMPORTANTE: Verifique se a coluna de nomes se chama exatamente 'NOME'
    # ou 'PRODUTO' no seu CSV. Use o nome exato abaixo:
    coluna_nome = 'Produto' # <--- AJUSTE AQUI

    df_ordenado = df_unificado.sort_values(by=coluna_nome)

    # 2. CAMINHO FINAL COM O NOME DO NOVO ARQUIVO (Sem barra no final também)
    caminho_final = '/content/drive/MyDrive/API_2LOG/estoque_unificado_final.csv'

    df_ordenado.to_csv(caminho_final, single_file=True, index=False)

    print(f"Sucesso! Arquivo gerado em: {caminho_final}")
else:
    print(f"Erro: Encontrados {len(existentes)} de 2 arquivos.")
    print("Verifique se os nomes abaixo estão na barra lateral do Colab:")
    for f in lista_especifica:
        print(f"{'✅' if os.path.exists(f) else '❌'} - {f}")
```
